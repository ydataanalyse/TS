```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('Documents/PROJETS PERSO/DATA SETS/Customers.csv')
df.head()
```

```python
#Informations sur la dataframe
df.info()
```

```python
#Renommer les colonnes

df1 = df.rename(columns={
    'CustomerID': 'ID_Client',
    'Gender': 'Genre',
    'Age': 'Âge',
    'Annual Income ($)': 'Revenu_annuel',
    'Spending Score (1-100)': 'Score_depenses',
    'Profession': 'Profession',
    'Work Experience': 'Experience_travail',
    'Family Size': 'Taille_famille'
})
```

```python
df1.head()
```

```python
#Transformer type "object" en "category"
df1['Genre'] = df1['Genre'].astype('category')
df1['Profession'] = df1['Profession'].astype('category')
df1.head(10)
```

```python
df1.info()
```

```python
#Vérifier les valeurs manquantes

df1.isnull().sum()
```

```python
#Suppression des valeurs manquantes
df2 = df1.dropna()
df2.isnull().sum()
```

```python
#Convertir la colonne 'Revenu_annuel'en euros
df2['Revenu_annuel'] = df2['Revenu_annuel'].apply(lambda x: x * 0.92)
df2.head()
```

```python
#tendances et distribution des données
df2.describe()
```

```python
df2.info()
```

```python
#Output
#df2.to_csv('TS.csv')
```

```python
from sklearn.cluster import KMeans

# Sélectionner les colonnes à utiliser pour le clustering
X = df2[['Âge', 'Revenu_annuel', 'Score_depenses', 'Experience_travail', 'Taille_famille']]

# Appliquer le clustering K-means avec K=5
kmeans = KMeans(n_clusters=5).fit(X)

# Ajouter les numéros de cluster à la dataframe
df2['cluster'] = kmeans.labels_
```
